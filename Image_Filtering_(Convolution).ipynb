```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
```

```python
img = Image.open('/content/sample_data/lenna.png')
img2 = Image.open('/content/sample_data/man.jpg')
display(img)
display(img2)
```

### Load the image using OpenCV

```python
im1 = cv2.imread('/content/sample_data/lenna.png')
#display(im1)
im2 = cv2.imread('/content/sample_data/man.jpg')
```

```python
im1.shape
```

```python
im2.shape
```

### Convert the RGB image into a grayscale image

```python
im1 = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
im1.shape
```

```python
im2 = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
im2.shape
```

### Define vertical and horizontal filter kernels

```python
v_kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])

h_kernel = np.array([

       [-1, -1, -1],

       [0, 0, 0],

       [1, 1, 1]

])
```

### Do the filtering

```python
v_image = cv2.filter2D(im1, -1, v_kernel)
h_image = cv2.filter2D(im1, -1, h_kernel)
```

```python
v_image2 = cv2.filter2D(im2, -1, v_kernel)
h_image2 = cv2.filter2D(im2, -1, h_kernel)
```

### Display results

```python
plt.imshow(im1, cmap='gray')
plt.title('Original Image')

plt.figure()
plt.imshow(v_image, cmap='gray')
plt.title('Vertical Image')

plt.figure()
plt.imshow(h_image, cmap='gray')
plt.title('Horizontal Image')
```
